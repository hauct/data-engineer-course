```python
# Láº¥y danh sÃ¡ch táº¥t cáº£ cÃ¡c báº£ng vÃ  schema
tables_df = db.read_sql("""
    SELECT 
        table_schema,
        table_name,
        table_type
    FROM information_schema.tables
    WHERE table_schema NOT IN ('information_schema', 'pg_catalog')
    ORDER BY table_schema, table_name
""")

print(f"\nðŸ“‹ Tá»•ng sá»‘ báº£ng: {len(tables_df)}")
print(tables_df)
```
