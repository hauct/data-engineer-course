## üìö SOLUTIONS

```python
# SOLUTION 1: Basic info
print(customers_df.info())
print("\n" + "="*50)
print(customers_df.describe())

# SOLUTION 2: Missing values
print(customers_df.isna().sum())

# SOLUTION 3: Unique countries
print(customers_df['country'].value_counts())

# SOLUTION 4: Filter Vietnam
vietnam_customers = customers_df[customers_df['country'] == 'Vietnam']

# SOLUTION 5: Uppercase names
customers_df['customer_name_upper'] = customers_df['customer_name'].str.upper()

# SOLUTION 6: Extract year/month
orders_df['order_date'] = pd.to_datetime(orders_df['order_date'])
orders_df['order_year'] = orders_df['order_date'].dt.year
orders_df['order_month'] = orders_df['order_date'].dt.month

# SOLUTION 7: Categorize orders
orders_df['order_category'] = pd.cut(
    orders_df['total_amount'],
    bins=[0, 500, 2000, float('inf')],
    labels=['Small', 'Medium', 'Large']
)

# SOLUTION 8: Merge
customer_orders = pd.merge(
    customers_df,
    orders_df,
    on='customer_id',
    how='inner'
)

# SOLUTION 9: Group by country
revenue_by_country = customer_orders.groupby('country')['total_amount'].sum().sort_values(ascending=False)

# SOLUTION 10: Customer analysis
customer_analysis = customer_orders.groupby('customer_id').agg({
    'order_id': 'count',
    'total_amount': ['sum', 'mean'],
    'order_date': ['min', 'max']
}).reset_index()

customer_analysis.columns = ['customer_id', 'total_orders', 'total_revenue', 'avg_order_value', 'first_order', 'last_order']
customer_analysis['days_since_last_order'] = (pd.Timestamp.now() - customer_analysis['last_order']).dt.days
customer_analysis['customer_lifetime_days'] = (customer_analysis['last_order'] - customer_analysis['first_order']).dt.days
```
