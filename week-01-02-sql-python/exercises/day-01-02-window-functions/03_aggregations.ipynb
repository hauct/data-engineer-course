## ðŸ“Š VISUALIZATION: Revenue trends

```python
# Monthly revenue trend
query = """
SELECT 
    DATE_TRUNC('month', order_date) as month,
    COUNT(*) as order_count,
    SUM(total_amount) as revenue
FROM analytics.orders
WHERE order_status = 'completed'
GROUP BY DATE_TRUNC('month', order_date)
ORDER BY month;
"""

df = run_query(query)
df['month'] = pd.to_datetime(df['month'])

fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 8))

# Revenue trend
ax1.plot(df['month'], df['revenue'], marker='o', linewidth=2)
ax1.set_title('Monthly Revenue Trend', fontsize=14, fontweight='bold')
ax1.set_ylabel('Revenue ($)')
ax1.grid(True, alpha=0.3)

# Order count trend
ax2.bar(df['month'], df['order_count'], color='orange', alpha=0.7)
ax2.set_title('Monthly Order Count', fontsize=14, fontweight='bold')
ax2.set_ylabel('Number of Orders')
ax2.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```
