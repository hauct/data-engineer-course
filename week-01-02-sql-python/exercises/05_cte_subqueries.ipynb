## ðŸŽ¯ EXERCISE 1: Simple CTE

**CÃ¢u há»i:** TÃ¬m customers cÃ³ revenue > average revenue

```python
# YOUR CODE HERE
query = """
WITH customer_revenue AS (
    SELECT 
        c.customer_id,
        c.customer_name,
        SUM(o.total_amount) as total_revenue
    FROM analytics.customers c
    INNER JOIN analytics.orders o ON c.customer_id = o.customer_id
    WHERE o.order_status = 'completed'
    GROUP BY c.customer_id, c.customer_name
),
avg_revenue AS (
    SELECT AVG(total_revenue) as avg_rev
    FROM customer_revenue
)
SELECT 
    cr.customer_name,
    cr.total_revenue,
    ar.avg_rev
FROM customer_revenue cr
CROSS JOIN avg_revenue ar
WHERE cr.total_revenue > ar.avg_rev
ORDER BY cr.total_revenue DESC;
"""

result = run_query(query)
result.head(20)
```
